```python
import random

tarot_deck = {
    1: "Шут (0) - Новые отношения, беззаботность, спонтанность. Будьте открыты к новому, но не теряйте голову.",
    2: "Маг (I) - Активное построение отношений, уверенность в себе, проявление инициативы.",
    3: "Жрица (II) - Интуиция в отношениях, тайные чувства, ожидание и наблюдение.",
    4: "Императрица (III) - Забота и плодородие в отношениях, создание уюта, проявление материнских чувств.",
    5: "Император (IV) - Стабильность и ответственность в отношениях, защита партнера, установление правил.",
    6: "Иерофант (V) - Традиционные ценности в отношениях, брак, следование общепринятым нормам.",
    7: "Влюбленные (VI) - Выбор в любви, гармония и страсть, принятие важного решения.",
    8: "Колесница (VII) - Движение вперед в отношениях, преодоление препятствий, контроль ситуации.",
    9: "Сила (VIII) - Внутренняя сила в отношениях, страсть и нежность, преодоление конфликтов.",
    10: "Отшельник (IX) - Самоанализ в отношениях, поиск себя, временное одиночество.",
    11: "Колесо Фортуны (X) - Перемены в отношениях, судьбоносные события, новые возможности.",
    12: "Справедливость (XI) - Честность и справедливость в отношениях, стремление к балансу и гармонии.",
    13: "Повешенный (XII) - Жертва в отношениях, переосмысление ценностей, готовность к изменениям.",
    14: "Смерть (XIII) - Завершение старого этапа в отношениях, трансформация и начало нового.",
    15: "Умеренность (XIV) - Гармония и баланс в отношениях, терпение и компромисс.",
    16: "Дьявол (XV) - Зависимость и искушения в отношениях, страхи и ограничения.",
    17: "Башня (XVI) - Разрушение старых иллюзий в отношениях, неожиданные перемены и кризис.",
    18: "Звезда (XVII) - Надежда и вера в лучшее в отношениях, исцеление и вдохновение.",
    19: "Луна (XVIII) - Неопределенность и иллюзии в отношениях, страхи и подсознательные чувства.",
    20: "Солнце (XIX) - Радость и счастье в отношениях, гармония и процветание.",
    21: "Суд (XX) - Переоценка отношений, прощение и возрождение чувств, шанс на новую жизнь.",
    22: "Мир (XXI) - Завершение цикла в отношениях, гармония и единство, полное удовлетворение."
}


card_statistics = {card: 0 for card in tarot_deck.keys()}


def update_card_statistics(card_number):
    """Обновляет статистику выбора карт."""
    if card_number in card_statistics:
        card_statistics[card_number] += 1


def display_card_statistics():
    """Отображает статистику выбора карт."""
    print("\nСтатистика выбора карт:")
    for card, count in card_statistics.items():
        print(f"Карта {card}: выбрана {count} раз(а)")


def analyze_popular_cards():
    """Анализирует и выводит самые популярные карты."""
    sorted_cards = sorted(card_statistics.items(), key=lambda x: x[1], reverse=True)
    

    top_3_cards = sorted_cards[:3]
    
    print("\nТоп-3 самых популярных карт:")
    for card, count in top_3_cards:
        print(f"Карта {card}: выбрана {count} раз(а)")

def draw_card(deck, card_number):
    """Возвращает описание карты по номеру."""
    return deck.get(card_number, "Карта не найдена")

def get_user_choice(available_cards, prompt):
    """Получает выбор пользователя с проверкой и запросом."""
    while True:
        try:
            choice = input(prompt)
            if not choice:
                print("Пожалуйста, сделайте выбор. Ввод не может быть пустым.")
                continue

            if not choice.isdigit():
                print("Пожалуйста, введите число.")
                continue

            choice = int(choice)

            if choice in available_cards:
                return choice
            else:
                print("Неверный ввод. Пожалуйста, выберите номер из предложенного списка.")
        except ValueError:
            print("Неверный ввод. Пожалуйста, введите целое число.")

def determine_outcome(card1_num, card2_num, card3_num, user_preference):
    """Определяет исход в зависимости от карт и предпочтений пользователя.
        card1: Прошлое/Опыт
        card2: Настоящее/Возможности
        card3: Будущее/Потенциал
    """

    first_card = draw_card(tarot_deck, card1_num)
    second_card = draw_card(tarot_deck, card2_num)
    third_card = draw_card(tarot_deck, card3_num)

    outcome = ""
    outcome += "-----------------------------\n"
    outcome += "Ваши предпочтения в отношениях:\n"
    if user_preference == "стабильность":
        outcome += "Для вас важна стабильность и предсказуемость в отношениях.\n"
    elif user_preference == "эмоциональная близость":
        outcome += "Для вас важна глубокая эмоциональная связь и понимание в отношениях.\n"
    elif user_preference == "доверие":
        outcome += "В отношениях для вас превыше всего доверие и честность.\n"
    elif user_preference == "страсть":
        outcome += "Вы ищете в отношениях яркие эмоции и страсть.\n"
    outcome += "-----------------------------\n"

    outcome += "Прошлое (Опыт): {first_card}\n".format(first_card=first_card)
    if card1_num == 1:  # Шут
        outcome += "  Возможно, в прошлом вы слишком легкомысленно относились к отношениям, бросаясь в них с головой, не задумываясь о последствиях. Это могло привести к разочарованиям и чувству нестабильности.\n"
    elif card1_num == 17:  # Башня
        outcome += "  В прошлом вы пережили болезненный разрыв или конфликт, который разрушил ваши иллюзии об отношениях. Это был период кризиса, который заставил вас пересмотреть свои ценности.\n"
    elif card1_num == 10:  # Отшельник
        outcome += "  В прошлом вы намеренно или вынужденно были одиноки. Возможно, вы взяли перерыв в отношениях, чтобы разобраться в себе и своих желаниях. Это время уединения могло быть как полезным, так и тягостным.\n"
    else:
        outcome += "  Ваш прошлый опыт в отношениях сформировал вас как личность и повлиял на ваши текущие взгляды на любовь и партнерство.\n"
    outcome += "-----------------------------\n"

    outcome += "Настоящее (Возможности): {second_card}\n".format(second_card=second_card)
    if card2_num == 20:  # Солнце
        outcome += "  Сейчас вы наслаждаетесь периодом счастья и гармонии в отношениях. Все складывается удачно, и вы чувствуете себя любимым и ценным. Это время радости и процветания.\n"
    elif card2_num == 7:  # Влюбленные
        outcome += "  Вы находитесь на перепутье и должны сделать важный выбор в любви. Возможно, вам предстоит решить, с кем вы хотите быть или какой путь выбрать в текущих отношениях. Доверьтесь своей интуиции, но будьте внимательны к последствиям.\n"
    elif card2_num == 16:  # Дьявол
        outcome += "  Вас одолевают искушения, зависимости или страхи, которые негативно влияют на ваши отношения. Возможно, вы чувствуете себя в ловушке или контролируемым. Необходимо осознать эти проблемы и принять меры для освобождения от них.\n"
    else:
        outcome += "  Перед вами открываются новые возможности для развития отношений. Важно быть открытым к новым знакомствам и не бояться проявлять инициативу.\n"
    outcome += "-----------------------------\n"

    outcome += "Будущее (Потенциал): {third_card}\n".format(third_card=third_card)
    if card3_num == 4:  # Императрица
        outcome += "  В будущем вас ждет стабильный и счастливый союз, наполненный заботой, любовью и плодородием (не обязательно в прямом смысле). Вы будете создавать уют и гармонию в доме, проявляя свою женственность и материнские качества.\n"
    elif card3_num == 18:  # Звезда
        outcome += "  Впереди вас ждет период надежды, исцеления и вдохновения в отношениях. Даже если сейчас вы переживаете трудные времена, верьте в лучшее. Эта карта предвещает восстановление и новые перспективы.\n"
    elif card3_num == 14:  # Смерть
        outcome += "  В будущем произойдет трансформация в ваших отношениях. Старое уйдет, чтобы освободить место для нового. Не бойтесь перемен, они необходимы для вашего роста и счастья.\n"
    else:
        outcome += "  Ваше будущее в отношениях зависит от ваших действий и решений в настоящем. Будьте осознанны и стремитесь к гармонии.\n"
    outcome += "-----------------------------\n"

    outcome += "Дополнительные интерпретации на основе комбинаций карт:\n"

    if card1_num == 17 and card2_num == 18:  # Башня и Звезда
        outcome += "  После разрушительного опыта (Башня) вы находите надежду и исцеление (Звезда). \n **Ситуация:** Вы пережили болезненный разрыв, но теперь начинаете восстанавливаться и верить в возможность новой любви. \n **Причина-следствие:** Разрыв отношений (Башня) привёл к периоду скорби, но теперь вы готовы к исцелению и новым возможностям (Звезда).\n"
    elif card1_num == 16 and card3_num == 22:  # Дьявол и Мир
        outcome += "  Освобождение от зависимостей (Дьявол) приведёт к гармоничным и удовлетворяющим отношениям (Мир). \n **Цель действия:** Работа над собой и избавление от негативных привычек (Дьявол) с целью достижения гармонии в отношениях (Мир). \n **Сфера жизни:** Личная жизнь, отношения.\n"
    elif card2_num == 7 and card3_num == 20:  # Влюбленные и Солнце
        outcome += "  Правильный выбор сейчас (Влюбленные) приведёт к радости и счастью в будущем (Солнце). \n **Интеллектуальная деятельность:** Анализ своих чувств и желаний (Влюбленные), чтобы принять правильное решение и построить счастливые отношения (Солнце).\n"
    elif card1_num == 10 and card2_num == 3:  # Отшельник и Жрица
        outcome += "Углубление в себя (Отшельник) и созерцание позволят увидеть в потоке рутины и будней проявления любви (Жрица). \n **Ситуация**: Анализ отношений при помощи интуиции, погружение в себя с целью осознать что-то важное.  \n"
    elif card1_num == 12 and card2_num == 6:  # Справедливость и Иерофант
        outcome += "Справедливость восторжествует, ваш союз признают, если Вы не опираетесь на предрассудки (Иерофант). \n **Причина и следствие**: соблюдение принципов честности и справедливости приводит к признанию отношений. \n **Сфера жизни**: семья, брак. \n"
    elif card1_num == 1 and card2_num == 7 and card3_num == 16:  # Шут, Влюбленные, Дьявол
        outcome += "Новое начало полно выбора, но будьте осторожны с искушениями и зависимостями. \n  **Цель действия**: Принятие решения в новых отношениях с учетом потенциальных опасностей. \n"
    elif card1_num == 3 and card2_num == 4 and card3_num == 15:  # Жрица, Императрица, Умеренность
        outcome += "Интуиция, забота и баланс ведут к гармоничным отношениям. \n **Характеристика человека**: Мудрая и заботливая личность, стремящаяся к гармонии.\n"
    elif card1_num == 17 and card2_num == 19 and card3_num == 10:  # Башня, Луна, Отшельник
        outcome += "После разрушения иллюзий важно уйти в себя и проанализировать свои страхи. \n **Интеллектуальная деятельность**: Анализ причин кризиса и работа со своими страхами. \n"
    elif card1_num == 14 and card2_num == 18 and card3_num == 20:  # Смерть, Звезда, Солнце
        outcome += "После завершения сложного этапа наступает исцеление и радость. \n **Ситуация**: Преодоление сложного этапа в отношениях и наступление периода счастья. \n"
    else:
        outcome += "  У каждой карты есть дополнительные слои смысла, исследуйте их.\n"
    outcome += "-----------------------------\n"

    return first_card, second_card, third_card, outcome

def main():
    print("Добро пожаловать в расклад Таро на тему любви!")
    print("Метод 3 карт дает информацию о вашем прошлом, настоящем и будущем. Сосредоточьтесь на своем вопросе. Выберите три карты. Карты ответят на ваш вопрос.")

    print("Что вы больше всего цените в партнере?")
    print("1. Стабильность")
    print("2. Эмоциональную близость")
    print("3. Доверие")
    print("4. Страсть")

    while True:
        preference_choice = input("Выберите номер ответа (1-4): ")
        if preference_choice == "1":
            user_preference = "стабильность"
            break
        elif preference_choice == "2":
            user_preference = "эмоциональная близость"
            break
        elif preference_choice == "3":
            user_preference = "доверие"
            break
        elif preference_choice == "4":
            user_preference = "страсть"
            break
        else:
            print("Неверный ввод. Пожалуйста, выберите номер от 1 до 4.")

    available_cards = list(tarot_deck.keys())

    # Запрос первой карты
    card1_num = get_user_choice(available_cards, "Выберите номер первой карты (Прошлое/Опыт) (1-22): ")
    card1 = draw_card(tarot_deck, card1_num)
    print(f"Первая карта: {card1_num} - {card1}")
    available_cards.remove(card1_num)
    update_card_statistics(card1_num)

    # Запрос второй карты
    card2_num = get_user_choice(available_cards, "Выберите номер второй карты (Настоящее/Возможности) (1-22): ")
    card2 = draw_card(tarot_deck, card2_num)
    print(f"Вторая карта: {card2_num} - {card2}")
    available_cards.remove(card2_num)
    update_card_statistics(card2_num)

    # Запрос третьей карты
    card3_num = get_user_choice(available_cards, "Выберите номер третьей карты (Будущее/Потенциал) (1-22): ")
    card3 = draw_card(tarot_deck, card3_num)
    print(f"Третья карта: {card3_num} - {card3}")
    update_card_statistics(card3_num)

    first_card, second_card, third_card, outcome = determine_outcome(card1_num, card2_num, card3_num, user_preference)

    print(f"\nПрогноз:\n{outcome}")

    # Отображение статистики выбора карт
    display_card_statistics()

    analyze_popular_cards()

if __name__ == "__main__":
    main()
```
